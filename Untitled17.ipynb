```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['cut']=le.fit_transform(data['cut'])
data['color']=le.fit_transform(data['color'])
data['clarity']=le.fit_transform(data['clarity'])
```

```python
data
```

```python
x = data.drop('price',axis=1)
y=data['price']
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mxc = MinMaxScaler()
```

```python
x_scl = mxc.fit_transform(x)
```

```python
x_scl
```

```python
x_scl.shape
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
```

```python
lr=LinearRegression ()
```

```python
lr.fit(x_scl,y)
preds = lr.predict(x_scl) 
print('r2 score ',r2_score(y,preds))
```

```python
from sklearn.preprocessing import PolynomialFeatures 
```

```python
poly = PolynomialFeatures (degree=3)
```

```python
x_poly = poly.fit_transform(x_scl)
```

```python
x_scaled_poly = mxc.fit_transform(x_poly)
```

```python
lr.fit(x_scaled_poly,y)
preds = lr.predict(x_scaled_poly) 
print('r2 score ',r2_score(y,preds))
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
train_preds = lr.predict(x_train) 
test_preds = lr.predict(x_test) 
print('r2 score ',r2_score(y_train,train_preds))
print('r2 score ',r2_score(y_test,test_preds))
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled_poly,y,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
train_preds = lr.predict(x_train) 
test_preds = lr.predict(x_test) 
print('r2 score ',r2_score(y_train,train_preds))
print('r2 score ',r2_score(y_test,test_preds))
```
